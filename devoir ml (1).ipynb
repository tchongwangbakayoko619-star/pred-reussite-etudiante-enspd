```python
# Fonction pour parser heures/temps en minutes
def parse_minutes(s):
    s = str(s).strip()
    if s in ["nan", ""]:
        return np.nan
    if "< 30 min" in s: return 15
    if "30min-1h" in s: return 45
    if "1h-2h" in s: return 90
    if "1h" in s: return 60
    if "2h" in s: return 120
    if "3h" in s: return 180
    if "4h" in s: return 240
    return np.nan

red["study_min"] = red["Heures d'etude personnelle par jour"].map(parse_minutes)
red["commute_min"] =red["Temps de trajet quotidien vers l'ecole"].map(parse_minutes)
```

```python
# Encodage ordinal
freq_map = {"Jamais":0, "Rarement":1, "Souvent":2, "Toujours":3}
inter_map = {"faible":0, "moyenne":1, "excellente":2}
bool_map = {"NON":0, "OUI":1}

for col in ["Acces au Wi-Fi de l'ecole","Acces aux equipements de labo","Acces a la bibliotheque"]:
    red[col+"_ord"] = red[col].map(freq_map)

red["salle_travail"] = red["une salle pour travailler ?"].map(bool_map)
red["ordi_perso"] =red["ordinateur personnel ?"].map(bool_map)
red["interaction_ord"] = red["Interaction avec les enseignant"].map(inter_map)
red["implication"] = red["Niveau d'implication dans les travaux de groupe"].astype(float)
# 5. Encodage des variables catgorielles (filire, niveau, logement) 
red = pd.get_dummies(red, columns=["Filiere","Niveau d'etude","Statut de logement"], drop_first=True)
# Normalisation / standardisation (optionnel pour modles linaires)  

# Colonnes numriques  normaliser/standardiser
num_cols = [
    "presence_pct","study_min","commute_min","implication","interaction_ord",
    "Acces au Wi-Fi de l'ecole_ord","Acces aux equipements de labo_ord","Acces a la bibliotheque_ord"
]

# Calcul de la moyenne et de l'cart-type sur chaque colonne
mean = red[num_cols].mean(axis=0)
std = red[num_cols].std(axis=0)

# Standardisation : (x - mean) / std
# On ajoute un petit epsilon pour viter la division par zro
epsilon = np.exp(-8)
red[num_cols] = (red[num_cols] - mean) / (std + epsilon)

# Vrification
print(red[num_cols].head())

#Vrification finale 
print(red.head()) 
print(red.info())
```

```python
# Fixer la graine pour reproductibilit 
np.random.seed(42) 
n = len(X)
```

```python
# Mlange des indices 
indices = np.random.permutation(n)
```

```python
# Split 80/20 
split = int(0.8 * n) 
X_train, X_test = X[indices[:split]], X[indices[split:]] 
Y_train, Y_test = Y[indices[:split]], Y[indices[split:]]
```

```python
# Ajouter une colonne de biais (utile si tu codes une rgression logistique "from scratch") 
X_train = np.hstack([np.ones((X_train.shape[0], 1)), X_train]) 
X_test = np.hstack([np.ones((X_test.shape[0], 1)), X_test])
```

```python
print("Shape X_train:", X_train.shape)
print("Shape X_test:", X_test.shape)
print("Shape Y_train:", Y_train.shape) 
print("Shape Y_test:", Y_test.shape)
```

```python
class LogisticRegression:
    def __init__(self, learningrate=0.01, iterations=500):
        self.lr = learningrate
        self.iterations = iterations
        self.weights = None
        self.bias = None
        self.costhistory = []  # on garde le cot  chaque itration

    # Fonction sigmode
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    # Entranement par descente de gradient
    def fit(self, X, y):
        n_samples, n_features = X.shape
        # Initialisation des paramtres
        self.weights = np.zeros(n_features)
        self.bias = 0

        for i in range(self.iterations):
            # Calcul des probabilits
            linear_model = np.dot(X, self.weights) + self.bias
            y_pred = self.sigmoid(linear_model)

            # Calcul du cot (log loss)
            cost = -(1/n_samples) * np.sum(y*np.log(y_pred+1e-8) + (1-y)*np.log(1-y_pred+1e-8))
            self.costhistory.append(cost)

            # Gradients
            dw = (1/n_samples) * np.dot(X.T, (y_pred - y))
            db = (1/n_samples) * np.sum(y_pred - y)

            # Mise  jour des paramtres
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

            # Affichage du cot tous les 100 pas
            if i % 100 == 0:
                print(f"Iteration {i}, Cost: {cost:.4f}")

    # Probabilits de prdiction
    def predict_proba(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return self.sigmoid(linear_model)

    # Conversion probabilit  classe (0 ou 1)
    def predict(self, X, threshold=0.5):
        proba = self.predict_proba(X)
        return (proba >= threshold).astype(int)
```

```python
model = LogisticRegression(learningrate = 0.01, iterations = 1000)
model.fit(X_train, Y_train)
```

```python
plt.plot(model.costhistory)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Courbe de convergence du cot")
plt.show()
```

```python
y_pred = model.predict(X_test)
accuracy = (y_pred == Y_test).mean()
print("Accuracy:", accuracy)
```

```python

# Matrice de confusion
def confusion_matrix_manual(y_true, y_pred):
    # Vrais positifs (TP) : prdits 1 et vrais 1
    TP = np.sum((y_true == 1) & (y_pred == 1))
    # Vrais ngatifs (TN) : prdits 0 et vrais 0
    TN = np.sum((y_true == 0) & (y_pred == 0))
    # Faux positifs (FP) : prdits 1 mais vrais 0
    FP = np.sum((y_true == 0) & (y_pred == 1))
    # Faux ngatifs (FN) : prdits 0 mais vrais 1
    FN = np.sum((y_true == 1) & (y_pred == 0))
    
    return np.array([[TN, FP],
                     [FN, TP]])

# Calcul des mtriques
def precision_manual(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    return TP / (TP + FP + 1e-8)


def recall_manual(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    return TP / (TP + FN + 1e-8)

def f1_score_manual(y_true, y_pred):
    prec = precision_manual(y_true, y_pred)
    rec = recall_manual(y_true, y_pred)
    return 2 * (prec * rec) / (prec + rec + 1e-8)

# Affichage
cm = confusion_matrix_manual(Y_test, y_pred)
print("Confusion Matrix:\n", cm)
print("Precision:", precision_manual(Y_test, y_pred))
print("Recall:", recall_manual(Y_test, y_pred))
print("F1-score:", f1_score_manual(Y_test, y_pred))
```
